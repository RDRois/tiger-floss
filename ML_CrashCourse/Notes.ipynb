## Notes on ML Concepts
### #Framing
**Labels**: The target of the prediction <br>
**Features**: Input variables used to train the model <br>
$${\{x_1, x_2, x_3,... x_N\}}$$
**Regression Model**: Predicts *continuous* values
> ex: What is the value of _?<br>
> ex: What is the probability that _?

### #Descending into ML: Linear Regression
**Classification Model**: Predicts *discrete* values <br>
**Linear Regression**: $$y'=b+w_1x_1$$ $$y'=b+w_1x_1+w_2x_2+w_3x_3$$
* $y'$ is the predicted label (a desired output).
* $b$ is the bias (the y-intercept), sometimes referred to as $w_0$.
* $w_1$ is the weight of feature 1. 
* $x_1$ is a feature (a known input).

### #Descending into ML: Training and Loss
**Mean square Error (MSE)**
$$MSE=\frac{1}{N}\sum_{(x,y)\in D}{(y-prediction(x))^2}$$
* $(x,y)$ is an example in which:
 * $x$ is the set of features.
 * $y$ is the example's label.
* $prediction(x)$ is a function of the weights and bias in combination with the set features $x$.
* D is a data set containing many labeled examples, which are $(x,y)$ pairs.
* N is the number of examples in D.

### #Reducting Loss: Gradient Descent
The **gradient** of a function is the vector of partial derivatives with respect to all the independent variables
if:
    $$f(x,y)=e^{2y}sin(x)$$
then:
    $$\nabla f(x,y)=\left(\frac{\delta f}{\delta x}(x,y),\frac{\delta f}{\delta y}(x,y)\right)=\left(e^{2y}cos(x),2e^{2y}sin(x)\right)$$
    
$\nabla f$   Points in the direction of greatest increase of the function. <br>
$-\nabla f$  Points in the direction of greatest decrease of the function.

## #Reducing Loss: Learning Rate
**Hyperparameters** are the knobbs that programmers tweak in ML algorithms. Tuning the learning rate allows for efficient discovery of loss minimum.
**Goldilocks** value is related to how flat the loss function is. "Just the right amount"

## #Reducing Loss: Stochastic Gradient Descent
**Full batch iteration** uses entire data set 
* slow
* redundancy
* enormous ~= large
**Stochastic gradient descent (SGD)** uses only a signle example(batch size=1) per iteration
* very noisy
* fast
* example chosen at random
**Mini-batch stochastic gradient descent (mini-batch SGD)** compromise between full-batch iteration and SGD. Uses typically between 10 and 1,000 examples, chosen at random. 
* reduces amount of noise 
* more efficient than full-batch

## #First steps with tensorflow
#### Hyperparameters
* **steps**, total number of training iterations. One step calculates the loss from *one* batch and uses that value to modify the model's weights *once*.
* **batch size**, the number of examples (chosen at random) for a single step. 
$$total\ number\ of\ trained\ examples = batch\ size*steps$$
* **periods**, controls the granularity of reporting. For example, if periods is set to 7 and steps is set to 70, then the exercise will output the loss value every 10 steps (or 7 times). Modifying periods does not alter what your models learns.

## Representation: Feature Engineering
* **Out-of-vocabulary (OOV)** - When categorical feature values are converted to numeric values and groupped into a catch-all "other" category  
    * map Charleston Road to 0
    * map North Shoreline Boulevard to 1
    * map Shorebird Way to 2
    * map Rengstorff Avenue to 3
    * map everything else (OOV) to 4
* **one-hot-encoding** - The length of the categorical feature in model is equal to th enumber of elements in the vocabulary.
* **Multi-hot encoding** - When multiple values are 1.
* **Sparse representation** - When there are only 1 or 2 elements in long binary vector that is of interest.
    * A Tensor that only stores nonzero elements.
    * **dense representation** is like entire dictionary.
    * Whereas a sparse representation will only include words in dictionary actually used.


## Representation: Cleaning Data
* **Z score** - tactic for scaling numerial data to a linear map
    * Scaling with Z scores means that most scaled values will be between -3 and +3, but a few values will be a little higher and lower than that range
$$scaledvalue = \frac{value-mean}{stddev}$$

* Handling extreme outliers
    * Raw data has a very long tail of outliers.
    * Take the log of every value: Tail is shorter, but still some large outlier values.
    * **Clip** the feature values at a max peak.
        * Leaves a single artifact at the outlier(max)
        * ex: roomsPerPerson = min(totalRooms / population, 4)
        * Clipping the feature value at 4.0 doesn't mean that we ignore all values greater than 4.0. <br>Rather, it means that all values that were greater than 4.0 now become 4.0.
    * **Quantile** Ensures the number of examples in each bin are equal - this removes the neeed to think about outliers
### Scrubbing
