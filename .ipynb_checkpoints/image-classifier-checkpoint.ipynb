# Run/Download docker image
You'll want to make sure your tensorflow image is working.
I had some trouble here because instructions on other sites kept pointing me in different directions. It was pretty annoying since I thought that with the login I would have access to any image. Like what is up with gcr.io
The error usually says something like:
>docker: Error response from daemon: manifest for gcr.io/tensorflow/tensorflow:latest-devel not found. <br>

This is bullsh*t. These 'suggested' images images other people used did NOT work for me:

>docker run -it gcr.io/tensorflow/tensorflow <br>
>docker run -it gcr.io/tensorflow/tensorflow:latest-devel <br>
>docker run -it gcr.io/tensorflow/tensorflow:latest-devel bash <br> 
>docker run -it b.gcr.io/tensorflow/tensorflow:latest-devel <br>
>docker run -it b.gcr.io/tensorflow/tensorflow:latest-devel bash <br>

This worked for me:
>docker run -it tensorflow/tensorflow:latest-devel bash<br>

You'll know it works because your terminal will show:
>root@XXXXXXXXXXX:~# <br>

The Xs are just a bunch of alpha numeric characters. iono if this makes it easier to read.
When you want to go back to a normal terminal, you can just type `exit` or you can press `CTRL+d`

### Check if have latest version of tensorflow inside the container
While inside the container, when looking into directories, a forward slash "/" needs to be used
>root@XXXXXXXXXXX:~# cd /tensorflow/ <br>

Next you'll want to update your tensorflow in the container
>root@XXXXXXXXXXX:/tensorflow# git pull <br>

This should start updating files in the container. 
If you've done this before it should show:
>Already up-to-date.

# Run retraining script
Alright finally we're here. Let's pray to lord flying spaghetti monster that this works. 
This is where most of the errors might show up. If the pathing is messed up, the files that are suppose to be generated by the retraining script will not appear in you tf-flyby folder. If there is a problem with the pathing, script will look like its working, but no files will show up in the tf-flyby folder.

So, check if your Docker container is in `/tf_files/` by typing
>root@XXXXXXXXXXX:/# ls /tf_files/
>ls: cannot access '/tf_files/': No such file or directory

Check if your Dcoker container is in the `/tf-flyby/` directory by typing
>root@XXXXXXXXXXX:/# ls /tf-flyby/
README.md  dogs  label_image.py  retrain.py  

Install tensorflow_hub. This is needed for retrain.py
>root@XXXXXXXXXXX:/# pip install tensorflow_hub 

Now you're probably ready to run the `retrain.py` file. You can copy paste the following.<br>
>root@XXXXXXXXXXX:/# python /tf-flyby/retrain.py --bottleneck_dir=/tf-flyby/bottlenecks --how_many_training_steps 500 --model_dir=/tf-flyby/inception --output_graph=/tf-flyby/retrained_graph.pb --output_labels=/tf-flyby/retrained_labels.txt --image_dir=/tf-flyby/dogs

If you get a "Not Found" Syntaxerror, it means there was a problem with you `retrain.py` file. Might want to try redownloading it checking for typos.

Below are some descriptions of the different parts needed to retrain.
* Run retrain script from /tf-flyby/ with the recently downloaded retrain file<br>
>root@XXXXXXXXXXX:~# python /tf-flyby/retrain.py \
* Cache the ouputs<br>
>--bottleneck_dir=/tf-flyby/bottlenecks \
* Run for 500 iterations, ideally ~4000+ <br>
>--how_many_training_steps 500 \
* Location to store trained model<br>
>--model_dir=/tf-flyby/inception \
* Visualization used in tensorflow<br>
>--output_graph=/tf-flyby/retrained_graph.pb \
* Output labels - same as traning folder name<br>
>--output_labels=/tf-flyby/retrained_labels.txt \
* Folder where images are stored <br>
>--image_dir=/tf-flyby/dogs<br>

**You'll know when it works lol**
Three things should appear in the tf-flyby folder. 
* bottlenecks folder
* retrained_graph.ph 
* retrained_labels.txt

# Test retrained model on new image
This last part is about taking the retrained model and applying it to a new image.
The output should list all the different dog types that were trained with percentages on which breed the new dog. Here is the [**documentation**](https://www.tensorflow.org/hub/tutorials/image_retraining) on image retraining from the tensorflow website. 

#### Get the new doggo image
Make sure you're not in the Docker container. 
>tf-flyby Home$ curl -LO https://camo.githubusercontent.com/934bfcb06c2b7b7b9d64a1391e5ad8909d28a929/687474703a2f2f7777772e707570636974792e636f6d2f5f6173736574732f696d616765732f6272656564732f33365f6d2e6a7067

<!--$ -->

Now rename because that name was obnouxiously long
>tf-flyby Home$ mv 687474703a2f2f7777772e707570636974792e636f6d2f5f6173736574732f696d616765732f6272656564732f33365f6d2e6a7067 doggo.jpg

<!--$ -->

#### Update/replace label_image.py
For whatever reason the `label_image.py` file included in most tutorials does not work. >:(<br>
In your local terminal, replace the `label_image.py` file and things should run more smoothely.

>tf-flyby Home$ curl -LO https://github.com/tensorflow/tensorflow/raw/master/tensorflow/examples/label_image/label_image.py

<!--$ -->

#### You're finally ready to test your new doggo.jog image!
Use the following in the Docker container, hopefully it works. Make sure there is no space after the "\", otherwise you'll get an "unrecogized arguments" error.

>root@XXXXXXXXXXX:~# python /tf-flyby/label_image.py \--graph=/tf-flyby/retrained_graph.pb \
--labels=/tf-flyby/retrained_labels.txt \--output_layer=final_result \--image=/tf-flyby/doggo.jpg \--input_layer=Placeholder
