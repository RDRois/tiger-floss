Here are some of the suggested images that did NOT work for me:

>Home$\$$　docker run -it gcr.io/tensorflow/tensorflow:latest-devel <br>
>Home$\$$　docker run -it gcr.io/tensorflow/tensorflow:latest-devel bash <br> 
>Home$\$$　docker run -it b.gcr.io/tensorflow/tensorflow:latest-devel <br>
>Home$\$$　docker run -it b.gcr.io/tensorflow/tensorflow:latest-devel bash <br>

This worked for me:
>Home$\$$ 　docker run -it tensorflow/tensorflow:latest-devel bash<br>

You'll know it works because your terminal will show:
>root@XXXXXXXXXXX:~#
The Xs are just a bunch of alpha numeric characters. iono if this makes it easier to read.
When you want to go back to a normal terminal, you can just type
exit 
or you can press CTRL+D

#### Check if have latest version of tensorflow inside the container
While inside the container, when looking into directories, a forward slash "/" needs to be used
>root@XXXXXXXXXXX:~# cd /tensorflow<br>
Next you'll want to update your tensorflow in the container
>root@XXXXXXXXXXX:/tensorflow# git pull<br>
This should start updating files in the container. 
If you've done this before it should show:
Already up-to-date.

* Run retrain script<br>
>python /tf_files/hub/examples/image_retraining/retrain.py \
* Cache the ouputs<br>
>\> --bottleneck_dir=/tf_files/tf-flyby/bottlenecks \<br>
* Run for 500 iterations<br>
>\> --how_many_training_steps 500 \<br>
* Location to store trained model<br>
>\> --model_dir=/tf_files/tf-flyby/inception \<br>
* Visualization used in tensorflow<br>
>\> --output_graph=/tf_files/tf-flyby/retrained_graph.pb \<br>
* Output labels - same as traning folder name<br>
>\> --output_labels=/tf_files/tf-flyby/retrained_labels.txt \<br>
* Folder where images are stored <br>
>\> --image_dir=/tf_files/tf-flyby/dogs<br>

#### Fix 1
**Include output and input operations**
"--output_layer=final_result"<br>
"--input_layer=Placeholder"
**Edit some parameters in label_image.py**
  input_height = 299
  input_width = 299
  input_mean = 0
  input_std = 255
  input_layer = "Mul" #Previously: "input"
  output_layer = "final_result" #Previously: "InceptionV3/Predictions/Reshape_1"
